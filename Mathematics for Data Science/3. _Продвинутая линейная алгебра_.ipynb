### Задание 1

Дан объект в $2D$ пространстве

````
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.show()
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````
# Уменьшим в два раза:
m1 = np.array([[0.5, 0, 0],
                [0, 0.5, 0],
                [0, 0, 1]])

scaled_A = A.dot(m1)
scaled_A
````

````
# Повернем на 130 градусов:
new_angle = 130
m2 = np.array([[np.cos(np.radians(new_angle)), -np.sin(np.radians(new_angle)), 0],
                            [np.sin(np.radians(new_angle)), np.cos(np.radians(new_angle)), 0],
                            [0, 0, 1]])

rotated_A = A.dot(m2)
rotated_A
````

````
# Отразим относительно прямой y=x:
m3 = np.array([[0, 1, 0],
                [1, 0, 0],
                [0, 0, 1]])

reflected_A = A.dot(m3)
reflected_A
````

````
# Визуализируем:
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(A[:, 0], A[:, 1])
axs[0, 0].set_title("Исходная матрица")

axs[0, 1].plot(scaled_A[:, 0], scaled_A[:, 1])
axs[0, 1].set_title("Матрица, уменьшенная вдвое")

axs[1, 0].plot(rotated_A[:, 0], rotated_A[:, 1])
axs[1, 0].set_title("Матрица, развернутая на 130 градусов")

axs[1, 1].plot(reflected_A[:, 0], reflected_A[:, 1])
axs[1, 1].set_title("Матрица, отраженная относительно прямой y=x")

# Установим единый масштаб
for ax in axs.flat:
    ax.set(xlim=(-200, 200), ylim=(-200, 200))

fig.suptitle('Галерея матриц')

plt.show()
````
