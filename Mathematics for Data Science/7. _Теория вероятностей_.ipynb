### Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

Задачу можно решить аналитически.
Вероятность купить продукцию высшего сорта можно рассчитать как сумму произведений процентного количества устройств с каждого предприятия на вероятность выбрать продукцию высшего сорта на этом предприятии:

(0.2 * 0.1) + (0.3 * 0.05) + (0.5 * 0.2) = 0.02 + 0.015 + 0.1 = 0.135

Таким образом, вероятность того, что случайно купленная новая продукция окажется высшего сорта, равна 0.135 или 13.5%.

```
# Проверим полученный результат
# Зададим вероятности выбора продукции конкретного предприятия
p1 = 0.2
p2 = 0.3
p3 = 0.5

# Зададим вероятности выбора продукции высшего сорта на конкретных предприятиях
p1_hq = 0.1
p2_hq = 0.05
p3_hq = 0.2

# Найдем вероятность того, что случайно купленная продукция окажется высшего сорта
p = (p1 * p1_hq) + (p2 * p2_hq) + (p3 * p3_hq)

print("Вероятность того, что случайно купленная новая продукция окажется высшего сорта: ", p)
```

```
# Решение с помощью моделирования случайных величин

import numpy as np
from scipy import stats

# Зададим вероятности выбора продукции конкретного предприятия
p1 = 0.2
p2 = 0.3
p3 = 0.5

# Зададим вероятности выбора продукции высшего сорта на конкретных предприятиях
p1_hq = 0.1
p2_hq = 0.05
p3_hq = 0.2

# Создадим дискретные случайные величины для выбора продукции
rv1 = stats.bernoulli(p1)
rv2 = stats.bernoulli(p2)
rv3 = stats.bernoulli(p3)

# Создадим дискретные случайные величины для выбора продукции высшего сорта
rv1_hq = stats.bernoulli(p1_hq)
rv2_hq = stats.bernoulli(p2_hq)
rv3_hq = stats.bernoulli(p3_hq)

# Моделируем выбор продукции конкретного предприятия с помощью дискретных случайных величин
product1 = rv1.rvs(100000)
product2 = rv2.rvs(100000)
product3 = rv3.rvs(100000)

# Моделируем выбор продукции высшего сорта на каждом предприятии
product1_hq = rv1_hq.rvs(100000)
product2_hq = rv2_hq.rvs(100000)
product3_hq = rv3_hq.rvs(100000)

p = np.mean((product1 * product1_hq) + (product2 * product2_hq) + (product3 * product3_hq))
print("Вероятность того, что случайно купленная новая продукция окажется высшего сорта: ", p)
```

Полученные результаты практически идентичны: вероятность того, что случайно купленная новая продукция окажется высшего сорта, составляет около 0.135 в первом коде и 0.13464 - во втором.

### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

Задачу можно решить аналитически. События А1, А2 и А3 попарно независимы, потому что наступление любого из этих событий не влияет на вероятность наступления двух других.

1) События А1 и А2: наступление события А1 не меняет вероятности наступления события А2.

2) События  А1 и А3: аналогично, наступление события А1 не меняет вероятности наступления события А3.

3) События  А2 и А3: наступление события А2 не меняет вероятности наступления события А3.

Таким образом, все три события попарно независимы.

Однако, если рассмотреть все три события вместе, они становятся зависимыми.

В том случае, если два из трех событий произошли (скажем, А1 и А2), то третье событие (А3) произойдет автоматически.

То есть, если монеты 1 и 2 упали одной стороной (событие А1) и монеты 2 и 3 упали одной стороной (событие А2),то автоматически  монеты 1 и 3 упали одной стороной (событие А3),поскольку обе упали так же, как и монета номер 2.

Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
```

```
# Параметры для нормального распределения
mean = [4,4,4,4,4]
std = [0.25, 0.50, 0.75, 1, 2]

# Визуализация
plt.figure(figsize=(10, 6))
for i in range(len(mean)):
    x = np.linspace(-4, 12, 100)
    y = norm.pdf(x, mean[i], std[i])
    plt.plot(x, y, label=f'mean={mean[i]}, std={std[i]}')
plt.title('Нормальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

Нормальное распределение:
   - Параметры: среднее значение (`mean`) и стандартное отклонение (`std`).
   - Увеличение среднего значения (`mean`) смещает график плотности вправо, уменьшение - влево.
   - Увеличение значения стандартного отклонения (`std`) делает график плотности широким и пологим, и, наоборот, уменьшение значения делает его узким и острым.

```
# Параметры для экспоненциального распределения
lambda_ = [0.25, 0.5, 0.75, 1, 2]

# Визуализация
plt.figure(figsize=(10, 6))
for i in range(len(lambda_)):
    x = np.linspace(0, 6, 100)
    y = expon.pdf(x, scale=1/lambda_[i])
    plt.plot(x, y, label=f'lambda={lambda_[i]}')
plt.title('Экспоненциальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

Экспоненциальное распределение:
   - Параметр: `lambda` (1/среднее значение)
   - Увеличение значения `lambda` делает график плотности более пологим на левой стороне и более плотным на правой стороне.
   - Уменьшение значения `lambda` делает график плотности более пологим на правой стороне и броее плотным на левой стороне.

```
# Параметры для распределения Стьюдента
df = [1, 2, 5, 12, 30]

# Визуализация
plt.figure(figsize=(10, 6))
for i in range(len(df)):
    x = np.linspace(-5, 5, 100)
    y = t.pdf(x, df[i])
    plt.plot(x, y, label=f'df={df[i]}')
plt.title('Распределение Стьюдента')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

Распределение Стьюдента:
   - Параметр: степени свободы (`df`)
   - Увеличение значения степеней свободы делает график плотности более заостренным вокруг среднего значения.
   - Уменьшение значения степеней свободы делает график плотности более выпуклым и плотным в хвостах.
