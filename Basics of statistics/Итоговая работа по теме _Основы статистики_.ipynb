**Решение:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('FPS_hw_x_df_x_taxi.xlsx')
df.head()
```

```python
# Формирование выборок
df_1 = df[df['offer_class_group'] == 'Economy']
df_2 = df[df['offer_class_group'] == 'Comfort']

print('Выборки имеют одинаковую длинну: ', len(df_1) == len(df_2))
```

```python
# Описательные статистики для класса поездок Economy
df_1.describe()
```

```python
# Описательные статистики для класса поездок Comfort
df_2.describe()
```

```python
# Визуализация плотности распределения дистанций поездок для классов Economy и Comfort
sns.distplot(df_1['distance_km'], 50)
sns.distplot(df_2['distance_km'], 50)
plt.title('Плотность распределения дистанций поездок для классов Economy и Comfort')
plt.legend(['Economy', 'Comfort'])
plt.show()
```

```python
# Визуализация распределения дистанций поездок для класса Economy
plt.plot(df_1['distance_km'])
plt.title('Плотность распределения дистанций поездок в классе Economy')
plt.legend(['Economy'])
plt.show()
```

```python
# Визуализация распределения дистанций поездок для класса Comfort
plt.plot(df_2['distance_km'])
plt.title('Плотность распределения дистанций поездок в классе Comfort')
plt.legend(['Comfort'])
plt.show()
```

Проверяем гипотезу: «На дальние поездки пользователи предпочитают Economy-классу более комфортные условия поездки в Comfort-классе.

Подключаем библиотеки статистических критериев и задаем уровень значимости

```python
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu

alpha = 0.05
```

Проведем Т-Тест (t-критерий Стьюдента)

```python
result = ttest_ind(df_1['distance_km'], df_2['distance_km'])
print('p-value: %.8f' % result.pvalue)

if (result.pvalue < alpha):
    print(' Отвергаем гипотезу')
else:
    print('Не отвергаем гипотезу')
```

Проведем тест: Манна — Уитни (U-критерий Манна — Уитни)

```python
result = mannwhitneyu(df_1['distance_km'], df_2['distance_km'])
print('p-value: %.8f' % result.pvalue)

if (result.pvalue < alpha):
    print('Отвергаем гипотезу')
else:
    print('Не отвергаем гипотезу')
```

Вывод: гипотезу отвергаем, статистически не выявлено предпочтений пользователей Comfort-классу при поездках на длинные дистанции.
