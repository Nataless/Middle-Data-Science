Создайте NumPy array с элементами от числа N до 0.
N не входит в массив. Для N = 10 результат будет array([9, 8, 7, 6, 5, 4, 3, 2, 1, 0])).

Важно: не удаляйте код ниже, а дополните его. Функция необходима для проверки вашего решения. Функция принимает целое число и возвращает np.array.



```python
import numpy as np

def my_array(N: int) -> np.array:
    x=np.arange(N - 1, -1, -1) if N > 0 else print(f'{N} - отрицательное число')
    return (x)

print(my_array(10))
print(my_array(15))
print(my_array(20))
```

Тренажер 2. Условие задачи
Создайте диагональную матрицу с элементами от N до 0 (N входит в массив).
Посчитайте сумму её значений на диагонали.

Важно: не удаляйте код ниже, а дополните его. Функция необходима для проверки вашего решения. Функция принимает целое число и возвращает целое число.

```python
import numpy as np

def sum_matrix(N: int) -> int:
    my_matrix = np.diagonal(np.diag(np.arange(N, -1, -1))).sum()
    return (my_matrix.sum())

print(sum_matrix(5))
print(sum_matrix(10))
print(sum_matrix(15))
```

```python
import numpy as np


def sys_lin_eq() -> np.array:
    a = np.array(
    [
        [4,2,1],
        [1,3,0],
        [0,5,4]
    ]
)
    b = np.array([4,12,-3])
    from numpy import linalg
    linalg.solve(a,b)

    return (linalg.solve(a,b))
```
