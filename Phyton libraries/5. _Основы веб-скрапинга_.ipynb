```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
```

Задача 1.

```python
def habr_search (query):
    habr_posts = pd.DataFrame()
    for q in query:
        URL = 'https://habr.com/ru/search/?'
        params = {
            'q': q
        }
        req = requests.get(URL, params=params)

        if req.status_code != 200:
            print(f"По данному запросу: {q} результатов не найдено!")
            continue

        soup = BeautifulSoup(req.text)

        news = soup.find_all('article', class_='tm-articles-list__item')

        for post in news:
            title = post.find('a','tm-title__link').text.strip()
            link = post.find('a','tm-title__link').get('href')
            date_original = post.find('a', 'tm-article-datetime-published tm-article-datetime-published_link').find('time').get('datetime')
            date = datetime.strptime(str(date_original), '%Y-%m-%dT%H:%M:%S.000Z')

            if 'megapost' in post.get('class', []):
                title += ' (megapost)'

            if 'remote-blog' in post.get('class', []):
                print(f"Упс...к сожалению, пост удален: {title}")
                continue

            row = {'Дата': date, 'Заголовок': title, 'Ссылка': 'https://habr.com' + link}
            habr_posts = pd.concat([habr_posts, pd.DataFrame([row])])


    return habr_posts.drop_duplicates().reset_index(drop=True)

res = habr_search(['python','анализ данных'])
res
```
