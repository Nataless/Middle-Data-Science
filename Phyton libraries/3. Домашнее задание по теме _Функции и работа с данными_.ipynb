```python
import pandas as pd
```

Задача 1.

```python
movies = pd.read_csv('./movies.csv')
ratings = pd.read_csv('./ratings.csv')
```

```python
added_movies = ratings.merge(movies, on='movieId', how='left')
added_movies[['movieId', 'title', 'rating']].head()
```

```python
def ratings_class(rating):
    if rating <= 2:
        return 'Low'
    elif rating <= 4:
        return 'Medium'
    else:
        return 'High'
```

```python
ratings['class'] = ratings['rating'].apply(ratings_class)
ratings.head()
```

```python
added_movies['class'] = added_movies['rating'].apply(ratings_class)
added_movies.sort_values('rating', ascending=False)
```

Задача 2.

```python
keywords = pd.read_csv('./keywords.csv')
keywords.head()
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_classification(keywords, geo_data):
    for keyword in keywords.split():
        for region, cities in geo_data.items():
            if keyword in cities:
                return region
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo_classification, geo_data=geo_data)
```

```python
keywords[keywords['region'] == 'Дальний Восток'].head()
```

Задача 3.

```python
import re
```

```python
def year_movie(movie):
    years = [i for i in range(1950, 2011)]
    try:
        m_year = re.search(r'\d{4}',movie).group(0)
    except:
        return 1900
    if int(m_year) in years:
        return m_year
    else:
        return 1900
```

```python
added_movies['year'] = added_movies['title'].apply(year_movie)
```

```python
added_movies.sort_values('title', ascending=True)
```

```python
added_movies[['year','rating']].groupby('year').mean().sort_values('rating', ascending=False).head()
```
