Задача 2.

```python
power= pd.read_csv('power.csv')
power = power[((power.country == 'Latvia') | (power.country == 'Lithuania') | (power.country == 'Estonia'))
      & (2005 <= power.year) & (power.year<=2010)
     & ((power.category == 4.0) | (power.category == 12.0) | (power.category == 21.0))
   & (power.quantity >= 0)].sort_values('quantity')
```

```python
power['quantity'].sum()
```
