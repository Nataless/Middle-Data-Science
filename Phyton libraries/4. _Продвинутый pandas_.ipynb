Задача 1.

```python
log = pd.read_csv('./visit_log.csv', sep=';')
```

```python
log.head(5)
```

```python
traffic_source = ('yandex', 'google', 'paid', 'email')
log['source_type'] = log['traffic_source']

log.loc[(log.traffic_source == traffic_source[0]) \
       | (log.traffic_source == traffic_source[1]), 'source_type'] = 'organic'

log.loc[(log.region == 'Russia') \
        & ((log.traffic_source == traffic_source[2]) \
           | (log.traffic_source == traffic_source[3])), 'source_type'] = 'ad'

log.loc[(log.region != 'Russia') \
        & ((log.traffic_source == traffic_source[2]) \
           | (log.traffic_source == traffic_source[3])), 'source_type'] = 'other'
```

Задача 3.

```python
ratings = pd.read_csv('./ratings.csv')
```

```python
ratings_of_top_100 = ratings.groupby(['userId']).filter(lambda x: len(x) > 100)
```

```python
ratings_of_top_100 = ratings_of_top_100.groupby(['userId']).agg({'timestamp': ['min', 'max']}).diff(axis=1)
```

```python
user_lifetime = ratings_of_top_100['timestamp'].reset_index()[['userId', 'max']].rename(columns={"max": "lifetime"})
user_lifetime.sort_values('lifetime', ascending=False).head()
```

Задача 4.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
total_revenue = (rzd.groupby('client_id').sum().reset_index()
           .merge(auto.groupby('client_id').sum().reset_index()
                  .merge(air.groupby('client_id').sum().reset_index(), how='outer')
                  , how='outer'
                 )
          ).fillna(0)
total_revenue.head()
```

```python
total_revenue_with_address = total_revenue.merge(client_base, how='right')
total_revenue_with_address.head()
```
