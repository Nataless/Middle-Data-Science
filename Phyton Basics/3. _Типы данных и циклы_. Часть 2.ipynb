Задание 1

через список:

```python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
unique_set = []
for number in ids.values():
    unique_set.extend(number)
print( set(unique_set))
```

через множества, поскольку они содержат только уникальное значение:



```python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
unique_set = set()
_ = [unique_set.update(x) for x in ids.values()]
print(f" {unique_set}")
```

задание 2


```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
queries_1 = []
for i in queries:
   queries_1.append(len(i.split(' ')))

dct = {}
for j in queries_1:
    if j in dct:
        dct[j] += 1
    else:
        dct[j] = 1
for item in sorted(dct):
    print('Поисковых запросов, содержащих', item, 'слов(а):', round((dct[item] * 100 / len(queries_1)), 2), '%')
```

задание 3


через обновление словаря:


```python
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24}
}
results_with_roi = {}

for row,values in results.items():
    results_with_roi.update({row:{'cost':values['cost'], 'revenue':values['revenue'], 'ROI':round((values['revenue'] / values['cost'] - 1) * 100, 2)}})

results_with_roi
```

через модуль pprint:

```python
from pprint import pprint
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}

for i in results.values():
    i['ROI'] = round(((i['revenue'] / i['cost'] - 1) * 100), 2)

pprint(results)
```

задание 6

```python
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},
    ]
}

my_recipe={}

person=int(input('Введите количество порций: \n'))

for item in cook_book.items():
    for ingridient in item[1]:
        ingridient_measure = ingridient['ingridient_name'].replace(' ','_') + ' ' + ingridient['measure']
        quantity = ingridient['quantity']
        if ingridient_measure in my_recipe:
            my_recipe[ingridient_measure] += quantity * person
        else:
            my_recipe.setdefault(ingridient_measure, quantity * person)

for ingridient, values in my_recipe.items():
    name = ingridient.split()
    print(name[0].capitalize().replace('_',' ') + ':', values, name[1])
```
