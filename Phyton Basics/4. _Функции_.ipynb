```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

```

```python
def numb_exist(number):
    """
    Функция проверяет, есть ли такой номер документа в перечне
    """
    for i in documents:
        if i['number'] == number:
            return True
    return False
```

```python
def holder(numb_doc):
    """
    Функция возвращает данные о владельце по номеру документа
    """
    holder_name = ''
    for row in documents:
        if row['number'] == numb_doc:
            holder_name = 'Владелец документа: ' + row['name']
        if holder_name == '':
            holder_name = 'Документ не найден в базе'
    return holder_name

```

```python
def doc_shelf(numb_doc):
    """
    Функция возвращает информацию о том, на какой полке хранится документ
    """
    shelf = ''
    for item, values in directories.items():
        if numb_doc in values:
            shelf = 'Документ хранится на полке: ' + item
    if shelf == '':
        shelf='Документ не найден в базе'
    return shelf
```

```python
def doc_list():
    """
    Функция возвращает список всех документов на полках
    """
    for i in documents:
        for item, values in directories.items():
            if i['number'] in values:
              print('№ ' + i['number'] + ', тип: ' + i['type'] + ', владелец: ' + i['name'] + ', полка хранения: ' + item)
```

```python
def list_shelf():
    """
    Функция возвращает список полок
    """
    listshelf = ''
    for item, values in directories.items():
        listshelf = listshelf + item + ','
    listshelf = listshelf[:-1] + '.'
    return listshelf
```

```python
def add_shelf(number_shelf):
    """
    Функция добавляет полку, если ее еще нет
    """
    count = 0
    for item, values in directories.items():
        if number_shelf in item:
            count = 1
    if count == 1:
        print('Такая полка уже существует. Текущий перечень полок: ' + list_shelf())
    else:
        directories[number_shelf]=[]
        print('Полка добавлена. Текущий перечень полок: '+ list_shelf())
```

```python
def del_shelf(number_shelf):
    """
    Функция удаляет полку, только если она пустая
    """
    count = 0
    for item, values in directories.items():
        if number_shelf in item:
            count = 1
            if values != []:
                count = 2
    if count == 1:
        del directories[number_shelf]
        print('Полка удалена. Текущий перечень полок: '+ list_shelf())
    elif count == 2:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + list_shelf())
    else:
        print('Такой полки не существует. Текущий перечень полок: ' + list_shelf())
```

```python
def main():
    """
    Итоговая функция автоматизации работы секретаря
    """
    a = 0
    while a < 1:
        command_input=input('Введите команду: \n')
        if command_input == 'p':
            number_input=input('Введите номер документа: \n')
            print(holder(number_input))
        elif command_input == 's':
            number_input=input('Введите номер документа: \n')
            print(doc_shelf(number_input))
        elif command_input == 'l':
            print (doc_list())
        elif command_input == 'ads':
            number_input=input('Введите номер полки: \n')
            add_shelf(number_input)
        elif command_input == 'ds':
            number_input=input('Введите номер полки: \n')
            del_shelf(number_input)
        elif command_input == 'q':
            break
        else:
            print('Ошибка при вводе команды')
```
