 ##**Домашнее задание по теме "Дашборды и принципы их построения"**

**Задание:**

Цель
Научиться создавать простой дашборд.

Важно: Данное задание с самопроверкой, оно не будет проверено экспертом, но после его выполнения вы получите вариант решения этой задачи от эксперта, и сможете сравнить ваш вариант с ним.
Выполнение этого задания позволит вам отработать теорию, которую вы изучили, и с лёгкостью справляться с подобными задачами на практике.
Умение анализировать требования заказчика и выделять из него задачи для визуализации - важный шаг к созданию визуализации, которой будут пользоваться.

Описание задания
Ваша задача создать дашборд по данным кейса.
Обратите внимание: вы можете использовать элементы визуализации, которые применялись для построения дашборда на занятии.
Дан набор данных по текущим и закрытым вакансиям за май и июнь (по состоянию на 09.06.2023). Описание набора данных и датасет доступны тутКак работать с файлом
Инструкция по выполнению задания
Ознакомьтесь с описанием кейса и сформулируйте цель дашборда и метрики, которые вам необходимы
Выделите из описания кейса проблемы клиента и задачи, которые ему нужно решить
Определите метрики для каждого решения
Определите цели, данные и визуализацию для метрик
Определите для себя цветовую схему, которой вы будете пользоваться
Сформируйте дашборд на основе предложенных рекомендаций
Продемонстрируйте историю на дашборде. Коротко на примере распишите, какие шаги делает рекрутёр в рамках дашборда, чтобы принять решение и решить свои проблемы
Дополнительно опишите все проблемы рекрутёра, которые были описаны в кейсе.
Формат выполнения: работа выполняется в Google Sheets / Excel.

**Рекомендации по построению дашборда в Google Sheets / Excel: **
Создайте отдельный лист в файле рядом с набором данных
Создайте сводную таблицу для каждой визуализации, которую вы определили (на основе исходных данных на листе, который вы создали)
На основе сводных таблиц постройте необходимые визуализации и примените к ним цветовые решения, которые вы определили
Соберите визуализации в дашборд
Дайте дашборду имя, проверьте все подписи к визуализациям
При необходимости добавьте срезы данных (фильтры) на основании источника данных.
Срезы будут действовать на все сводные таблицы и визуализации, которые вы создали на своем листе

Результат выполненного задания:
для корректной сдачи домашнего задания необходимо в личном кабинете прикреплять ссылку на готовое решение, открыв доступ на просмотр и комментирование;
после того, как вы прикрепите ссылку на своё решение в личном кабинете, вам станет доступен пример решения, который подготовил эксперт.
