##**Домашнее задание по теме «Виды визуализаций. Выбор правильной визуализации»**

**Задание**

Цель
Научиться выбирать правильную визуализацию под конкретную задачу с учетом типа данных

Важно:
Данное задание с самопроверкой, оно не будет проверено экспертом, но после выполнения вы получите вариант решения этой задачи от эксперта, и сможете сравнить ваш вариант с ним.
Выполнение этого задания позволит вам отработать теорию, которую вы изучили, и с лёгкостью справляться с подобными задачами на практике. Умение правильно подбирать визуализацию позволит вам улучшить качество визуализаций, которые вы создаете. Ведь 60% успеха визуализации зависит от правильно подобранного графика.

Контекст
Дан набор данных по результатам исследования о выплатах специалистам в области Анализа данных, проведенное в США.
Данные доступны тут Как работать с файлом

Перечень полей:
work_year — год (2022, 2023)
experience level — уровень экспертизы (EX - Executive level; SE - Senior; MI - Mid/Intermediate level; EN - Entry level)
employment type — тип трудоустройства (FT - full-time job; PT - part-time job; FL - Freelancer; CT - Contractor)
job_title — наименование позиции
salary_in_usd — заработная плата за год
remote_ratio — возможность работать удалённо (100 - удалённо, 0 - в офисе, 50 - гибридный формат)
company_size — размер компании (L - крупный бизнес, M - средний бизнес, S - малый бизнес)

Задание
Выполните любые 3 задачи из предложенных ниже (задачи расположены по увеличению уровня сложности):
Задача 1. Показать как изменилась заработная плата за год (с 2022 по 2023) по специальностям
Задача 2. Сравнить медианные заработные платы по специальностям
Задача 3. Сравнить количество вакансий разного уровня экспертизы
Задача 4. Показать распределение заработных плат
Задача 5. Сравнить распределение заработных плат по уровням специалистов

Инструкция по выполнению задания
Выберите любые 3 задачи из предложенных.
Для каждой задачи укажите, какой тип задач вы решаете: сравнение, пропорция, распределение.
Выберите необходимые данные, которые нужны для визуализации и выполните необходимые расчеты.
Подберите и постройте подходящую визуализацию, исходя из выбранной задачи и типа данных, которые для нее необходимы
Под каждым графиком сделайте вывод о том, ЧТО он показывает.

Формат выполнения:
работа выполняется в Google Sheets / Excel.

Результат выполненного задания:
для корректной сдачи домашнего задания необходимо в личном кабинете прикреплять ссылку на готовое решение, открыв доступ на просмотр и комментирование;
после того, как вы прикрепите ссылку на своё решение в личном кабинете, вам станет доступен пример решения, который подготовил эксперт.

Чек-лист самопроверки
Перед сдачей домашнего задания убедитесь, что вы:
прописали задачи, которые вы решаете
дали графику наименование (если у вас получилось более одного графика при решении задачи - это может быть одно наименование)
выделили наименования метрик на осях и единицы измерения
построили графики в соответствии с рассмотренными на занятии правилами
оформили подписи данных (если они есть) так, что они не наезжают друг на друга
применили корректную цветовую схему (если её использовали, а если у вас получилось более одного графика при решении задача - у вас общая цветовая схема для всех графиков)
использовали только один шрифт и не более 3-х размеров шрифта
прописали основной вывод о том, что отображено на графике для каждой задачи

Успехов в выполнении задания!
