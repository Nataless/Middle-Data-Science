Посчитаем базовые статистики для непрерывных случайных величин:

```python
def stat(column):
    print(f'Среднее значение столбца {column.name} - {column.mean()}')
    print(f'Медианное значение столбца {column.name} - {column.median()}')
    print(f'Модальное значение столбца {column.name} - {column.mode()[0]}')
    print(f'Стандартное отклонение столбца {column.name} - {column.std()}')
    print(f'Дисперсия столбца {column.name} - {column.var()}')
```

```python
stat(df['rectal temperature'])
```

```python
stat(df['respiratory rate'])
```

```python
stat(df['pulse'])
```

Посчитаем выбросы через межквартильный размах:

```python
def find_outlier(df, param):
    q1 = df[param].quantile(0.25)
    q3 = df[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    df2= df[~df[param].between(lower_bound, upper_bound, inclusive=True)]
    return df2.loc[df2[param].notnull()]
```

```python
q1 = df['rectal temperature'].quantile(0.25)
q3 = df['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal temperature'].between(lower_bound, upper_bound)]
remove_outliers
```

При норме данного показателя - 37.8. в выбросы попали как высокие, так и низкие значения.Высокие значения, согласно описанию данных, могут быть связаны с инфекцией, а низкие - с последними стадиями шока.

```python
q1 = df['respiratory rate'].quantile(0.25)
q3 = df['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['respiratory rate'].between(lower_bound, upper_bound)]
remove_outliers
```

В выборку попали лошади с очень высокими значениями частоты дыхания, которые значительно превышают норму.

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound)]
remove_outliers
```

В выбросы попали очень высокие значения пульса в кластере взрослых лошадей, которые могут быть вызваны болезненными поражениями или цикличным шоком. Этот факт косвенно подтверждают соответсвенно высокие значения
respiratory_rate, rectal temperature и pain.
Из 5 лошадей выжила только та, что была прооперирована. Следовательно, можно сделать вывод: если у взрослой лошади значительно повышены данные показатели (в выбросах по значению), то ей необходима операция, иначе шансов на выживаемость нет.

Заполним пропуски модой для категориальных атрибутов, в который есть пропуски:

```python
df_new['surgery?'] = df_new['surgery?'].fillna(df['surgery?'].mode()[0])
df_new['temperature of extremities'] = df_new['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0])
df_new['pain'] = df_new['pain'].fillna(df['pain'].mode()[0])
df_new['outcome'] = df_new['outcome'].fillna(df['outcome'].mode()[0])
df_new
```

```python
df_new[['rectal temperature', 'pulse', 'respiratory rate']]= df_new[['rectal temperature', 'pulse', 'respiratory rate']].fillna(df[['rectal temperature', 'pulse', 'respiratory rate']].mean()).round(1)
df_new
```
